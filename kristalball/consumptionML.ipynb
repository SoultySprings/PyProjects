```python
dfAnderson = df[df['Bar Name'].str.contains("Anderson")].sort_values(["Alcohol Type"], ascending=True).copy()
dfBrown = df[df['Bar Name'].str.contains("Brown")].sort_values(["Alcohol Type"], ascending=True).copy()
dfSmith = df[df['Bar Name'].str.contains("Smith")].sort_values(["Alcohol Type"], ascending=True).copy()
dfTaylor = df[df['Bar Name'].str.contains("Taylor")].sort_values(["Alcohol Type"], ascending=True).copy()
dfThomas = df[df['Bar Name'].str.contains("Thomas")].sort_values(["Alcohol Type"], ascending=True).copy()
```

```python
figAndyBrPur = px.hist_frame(dfAnderson,
                     x='Brand Name',
                     y='Purchase (ml)',
                     marginal='box',
                     color = 'Brand Name',
                     color_discrete_sequence=px.colors.qualitative.G10,
                     title='Brands').update_layout(bargap=0.1)
figBrownBrPur = px.hist_frame(dfBrown, x='Brand Name', y='Purchase (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figSmithBrPur = px.hist_frame(dfSmith, x='Brand Name', y='Purchase (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figTayBrPur = px.hist_frame(dfTaylor, x='Brand Name', y='Purchase (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figThomBrPur = px.hist_frame(dfThomas, x='Brand Name', y='Purchase (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)


figAndyBrPur.show()
figBrownBrPur.show()
figSmithBrPur.show()
figTayBrPur.show()
figThomBrPur.show()

figAndyPurCons = px.make_subplots(rows=1, cols=2)
```

```python
figAndyBrCons = px.hist_frame(dfAnderson,
                     x='Brand Name',
                     y='Consumed (ml)',
                     marginal='box',
                     color = 'Brand Name',
                     color_discrete_sequence=px.colors.qualitative.G10,
                     title='Brands').update_layout(bargap=0.1)
figBrownBrCons = px.hist_frame(dfBrown, x='Brand Name', y='Consumed (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figSmithBrCons = px.hist_frame(dfSmith, x='Brand Name', y='Consumed (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figTayBrCons = px.hist_frame(dfTaylor, x='Brand Name', y='Consumed (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)
figThomBrCons = px.hist_frame(dfThomas, x='Brand Name', y='Consumed (ml)', marginal='box',
                     color = 'Brand Name', color_discrete_sequence=px.colors.qualitative.G10, title='Brands').update_layout(bargap=0.1)

from plotly.subplots import make_subplots

figAndyBrCons.show()
figBrownBrCons.show()
figSmithBrCons.show()
figTayBrCons.show()
figThomBrCons.show()
```
